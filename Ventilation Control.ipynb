**Мостът (pons)** представлява базално издута част на мозъчния ствол, която се разполага под средния мозък, над гръбначния и пред малкия мозък. Тази част от мозъчния ствол се нарича още и мост на Варолий (pons Varolii) - по името на италианския анатом и хирург Costanzo Varolio (1543-75). Мостът има функционална връзка с малкия мозък и включва нервни пътища, които предават импулси от крайномозъчната кора до малкия и гръбначния мозък, и такива, които носят сетивни сигнали към таламуса.

**Продълговатият мозък (medulla oblongata)**, е част от мозъчния ствол на главния мозък и се разполага най-ниско в черепната кухина - отпред и частично под малкия мозък. Надолу се продължава в гръбначния мозък, с който имат сходно устройство и функции. За граница между продълговатия и гръбначния мозък се приема областта, от която излиза първия гръбначномозъчен нерв, а за граница с надлежащия мост (pons) отпред служи дълбока напречна бразда - sulcus bulbopontinus, от която изхождат 6-ти, 7-ми и 8-ми черепномозъчни нерви, а отзад - striae medullares.

Функцията на продълговатия мозък е свързана с регулация на редица вегетативни (автономни) процеси като сърдечна дейност, дишане, кръвно налягане. В него са разположени вегетативни центрове, които представляват интегрална част на рефлекси като кихане, кашляне, повръщане и бозаене.

Източник - https://medpedia.framar.bg/

При високи нива на $CO_2$ в кръвоносния съд, хеморецепторите не могат да отделят произведения въглероден двуокис в кръвта и той се натрупва около тях. Те реагират на повишените нива на $CO_2$ като изпращат сигнал по невротрансмитерите към дихателните центрове и по този начин скоростта на дишане се контролира физиологично.

Нека означим с:

$\tau$ - интервала на дишане

$V_n = V(t + n\tau)$ - вентилационния обем, който зависи от:

$C_{n-1} = C(t + (n-1)\tau)$ - количеството $CO_2$ в предходния момент

$S(C_n)$ - чувствителността на хеморецепторите към $CO_2$ в кръвта

$L(V_n, C_n)$ - количеството $CO_2$ вентилирано през белите дробове

$m$ - постоянния приток на $CO_2$ като продукт от нормалния метаболизъм


Тогава можем да запишем следната система:

$\begin{cases}
   C_{n+1} = C_n - L(V_n, C_n) + m\\
   V_{n+1} = S(C_n)
 \end{cases}$

## Линейна формула за $L(V_n, C_n)$

Допускаме, че количеството $CO_2$ вентилирано през белите дробове ($L(V_n, C_n)$) е пропорционално на вентилационния обем ($V_n$) и не зависи от количеството на $CO_2$ в сегашния момент ($C_n$), т.е $\mathbf{L(V_n, C_n) = \beta V_n}$.

Допускаме също така, че вентилационният обем в следходния момент ($V_{n+1}$) е пропорционален на количеството на $CO_2$ в сегашния момент ($C_n$), т.е $\mathbf{V_{n+1} = \alpha C_n}$.

Тогава получаваме следната система:
$\begin{cases}
   C_{n+1} = C_n - \beta V_n + m\\
   V_{n+1} = \alpha C_n
 \end{cases} \Leftrightarrow 
 \begin{cases}
   C_{n+1} = C_n - \beta V_n + m\\
   V_{n} = \alpha C_{n-1}
 \end{cases}
 $

Като заместим $V_n$ в уравнението за $C_{n+1}$ получаваме едно уравнение за $C$:

$\mathbf{C_{n+1} = C_n - \alpha \beta C_{n-1} + m}$

- Ако $m = 0$, уравнението е хомогенно.
  Разглеждаме характеристичния му полином:
  
  $x^2 = x - \alpha \beta$

  Той има корени $x_{1;2} = \frac{1 \pm \sqrt{1 - 4 \alpha \beta}}{2}$

  Тогава общото решение на хомогенното уравнение е:
  $A(n) = C_1 (\frac{1 + \sqrt{1 - 4 \alpha \beta}}{2})^n + C_2 (\frac{1 - \sqrt{1 - 4 \alpha \beta}}{2})^n $, където $C_1$ и $C_2$ са константи зависещи от началните условия.


- Ако $m \neq 0$, уравнение е нехомогенно.
  Търсим частно решение от вида $B(n) = k, k - const$. Щом $B(n)$ е решение, тогава е изпълнено:

  $k = k - \alpha \beta k + m$, от където $k = \frac{m}{\alpha \beta}$.
  
  Следователно, общото решение на нехомогенното уравнение е:
  $C_n = A(n) + B(n) = C_1 (\frac{1 + \sqrt{1 - 4 \alpha \beta}}{2})^n + C_2 (\frac{1 - \sqrt{1 - 4 \alpha \beta}}{2})^n + \frac{m}{\alpha \beta}$.

- Нека $4\alpha \beta < 1$. Разглеждаме модул от собствените стойности $\mid\frac{1 + \sqrt{1 - 4 \alpha \beta}}{2}\mid$ и $\mid\frac{1 - \sqrt{1 - 4 \alpha \beta}}{2}\mid$.

  От $\sqrt{1 - 4 \alpha \beta} < 1 \implies$ и двете собствени стойности са $< 1$.
  
  Следователно, нивото на въглероден двуокис ще се стабилизира в точка $\bar{C} = \frac{m}{\alpha \beta}$.
  
  Вентилационният обем в тази точка ще бъде $\bar{V} = \alpha \bar{C} = \alpha \frac{m}{\alpha \beta} = \frac{m}{\beta}$.

  Да видим:

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def linear_model_formula(iterations, alpha, beta, m, c1=1, c2=1):
    result = []
    for n in range(iterations):
        c_n = c1 * ((1 + math.sqrt(1 - 4*alpha*beta)) / 2)**n + c1 * ((1 - math.sqrt(1 - 4*alpha*beta)) / 2)**n + (m / (alpha * beta))
        result.append([c_n, alpha*c_n])
    return result


def plot_model(result):
    ll = plt.plot(result)
    xl = plt.xlabel('Време')
    plt.legend(ll, ["Количество $CO_2$", "Вентилационен обем"])
    plt.show()
```

```python
plot_model(linear_model_formula(100, 0.5, 0.3, 1.2)) # iterations=100, alpha=0.5, beta=0.3, m=1.2, c1=c2=1
```

Виждаме, че количеството въглероден двуокис ($C_n$) се стабилизира в точка $\frac{m}{\alpha \beta} = \frac{1.2}{0.5*0.3} = 8$, a вентилационният обем в $\frac{m}{\beta} = \frac{1.2}{0.3} = 4 $.

Дори и да променим стойностите на константите $C_1$ и $C_2$, зависещи от началните условия, $C_n$ и $V_n$ отново ще се стабилизират в същите точки:

```python
plot_model(linear_model_formula(100, 0.5, 0.3, 1.2, c1=3, c2=2)) # iterations=100, alpha=0.5, beta=0.3, m=1.2, c1=3, c2=2
```

Да проверим дали затворената формула води до същите резултати като рекурентната:

```python
def linear_model(iterations, alpha, beta, m, c0=1, c1=1, v0=1):
    result = [[c0, v0], [c1, alpha * c0]]
    c_previous = c0
    c_current = c1
    for _ in range(iterations):
        c_next = c_current - alpha * beta * c_previous + m
        v_next = alpha * c_current
        result.append([c_next, v_next])
        c_previous = c_current
        c_current = c_next
    return result
```

```python
plot_model(linear_model(100, 0.5, 0.3, 1.2, c0=10, c1=9, v0=5.0))
plot_model(linear_model_formula(100, 0.5, 0.3, 1.2, c1=1, c2=1))
```

- Нека $4\alpha \beta > 1$. Тези стойности не са в дефиниционното множество на затворената формула, тъй като $\sqrt{1 - 4 \alpha \beta}$ няма да бъде дефинирано за реални стойности на параметрите $\alpha$ и $\beta$.

  Но да погледнем какво се случва с рекурентния модел при такива стойности:

```python
plot_model(linear_model(100, 0.9, 0.9, 1.2, c0=5, c1=3, v0=3)) # alpha*beta = 0.81
```

```python
plot_model(linear_model(100, 1, 1, 1.2, c0=5, c1=3)) # alpha*beta = 1
```

```python
plot_model(linear_model(100, 1.1, 1.1, 1.2, c0=5, c1=3)) # alpha*beta = 1.21
```

```python
plot_model(linear_model(50, 1.5, 1.5, 1.2, c0=5, c1=3)) # alpha*beta = 2.25
```

Виждаме, че при достатъчно голямо $\alpha \beta$ колебанията растат и при това с големи темпове. В последния пример, където $\alpha \beta = 2.25$ стойностите достигат разлика от $10^9$. 

## Нелинейна формула за $L(V_n, C_n)$

Модифицираме модела така, че количеството $CO_2$ вентилирано през белите дробове ($L(V_n, C_n)$) да зависи от количеството на $CO_2$ в сегашния момент, т.е. $\mathbf{L(V_n, C_n) = \beta V_n C_n}$. Запазваме формулата за вентилационния обем (да е пропорционален на $C_n$, т.е. $\mathbf{V_{n+1} = \alpha C_n}$).

Тогава системата

$\begin{cases}
   C_{n+1} = C_n - L(V_n, C_n) + m\\
   V_{n+1} = S(C_n)
 \end{cases}$

става

$\begin{cases}
   C_{n+1} = C_n - \beta V_n C_n + m\\
   V_{n+1} = \alpha C_{n}
 \end{cases}\Leftrightarrow 
 \begin{cases}
   C_{n+1} = C_n - \beta V_n C_n + m\\
   V_{n} = \alpha C_{n-1}
 \end{cases}$

Като заместим $V_n$ в уравнението за $C_{n+1}$ получаваме едно уравнение за $C$:

$\mathbf{C_{n+1} = C_n - \alpha \beta C_n C_{n-1} + m}$

Търсим равновесна точка, т.е. точка за която $C_{i+1} = C_i = \bar{C}, \quad \forall i > i' \iff \bar{C} = f(\bar{C})$. Тогава:

$f(C) = \bar{C} - \alpha \beta \bar{C}^2 + m$

$\bar{C} = f(\bar{C}) \Leftrightarrow \bar{C} = \bar{C} - \alpha \beta \bar{C}^2 + m \implies \alpha \beta \bar{C}^2 = m \implies \bar{C}^2 = \frac{m}{\alpha \beta} \implies \bar{C} = \sqrt{\frac{m}{\alpha \beta}}$

Изследваме равновесната точка за устойчивост. (За да е устойчива равновесната точка, трябва $\mid f'( \bar{C})\mid < 1$.)

$\mid f'( \bar{C})\mid = \mid \bar{C} - \alpha \beta \bar{C}^2 + m \mid \quad \Leftrightarrow \quad \mid f'( \bar{C})\mid = \mid - 2 \alpha \beta \mid$

Следователно, равновесната точка

$\bar{C} = \sqrt{\frac{m}{\alpha \beta}} \text{  е  }
   \begin{cases}
     \text{устойчива} & \quad \text{за } \alpha \beta < \frac{1}{2}\\
     \text{неустойчива} & \quad \text{за } \alpha \beta > \frac{1}{2}
   \end{cases}$

и съответно $\bar{V} = \alpha \bar{C}$.

Да изследваме модела при тези условия:

```python
def nonlinear_model(iterations, alpha, beta, m, c0=1, c1=1, v0=1):
    result = [[c0, v0], [c1, alpha * c0]]
    c_previous = c0
    c_current = c1
    for _ in range(iterations):
        c_next = c_current - alpha * beta * c_current * c_previous + m
        v_next = alpha * c_current
        result.append([c_next, v_next])
        c_previous = c_current
        c_current = c_next
    return result
```

При $\alpha \beta < \frac{1}{2}$:

```python
plot_model(nonlinear_model(100, 0.4, 0.4, 4)) # alpha * beta = 0.16, m=4
```

```python
plot_model(nonlinear_model(100, 0.4, 0.4, 2)) # alpha * beta = 0.16, m=2
```

При $\alpha \beta > \frac{1}{2}$:

```python
plot_model(nonlinear_model(150, 0.8, 0.8, 1.2)) # alpha * beta = 0.64, m=1.2
```

```python
plot_model(nonlinear_model(150, 0.8, 0.8, 2)) # alpha * beta = 0.64, m=2
```

```python
plot_model(nonlinear_model(100, 0.9, 0.9, 1.2)) # alpha * beta = 0.81
```

```python
plot_model(nonlinear_model(100, 1, 1, 1.2)) # alpha * beta = 1
```

```python
plot_model(nonlinear_model(100, 1.1, 1, 1.2)) # alpha * beta = 1.1
```

```python
plot_model(nonlinear_model(100, 1.1, 1.1, 1.2)) # alpha * beta = 1.21
```

От примерните графики се вижда, че при $\alpha \beta > \frac{1}{2}$, има колебания в $V_n$ и $C_n$.

Нещо повече, отново както при линейния модел се наблюдава, че колебанията се увеличават заедно с нарастването на $\alpha \beta$ и то с бързи темпове. В последния пример при $\alpha \beta = 1.21$ възникват стойности, прекалено дълги за изчисление.

## Сигмоидна функция за $S(C)$

Mackey и Glass (1977, 1978) разглеждат сигмоидна функция за чувствителността на хеморецепторите,

$S(C) = \frac{C^l}{K^l + C^l}$, с параметри $l \in \mathbb{N}, K \in \mathbb{R}$.

Нека $V_{n+1} = V_{max}S(C_n)$, където $V_{max}$ е максималният обем.

За $L(V_n, C_n)$ запазваме нелинейната зависимост $L(V_n, C_n) = \beta V_n C_n$.

Тогава системата

$\begin{cases}
   C_{n+1} = C_n - L(V_n, C_n) + m\\
   V_{n+1} = S(C_n)
 \end{cases}$

става

$\begin{cases}
   C_{n+1} = C_n - \beta V_n C_n + m\\
   V_{n+1} = V_{max}\frac{C_n^l}{K^l + C_n^l}
 \end{cases}\Leftrightarrow
 \begin{cases}
   C_{n+1} = C_n - \beta V_n C_n + m\\
   V_{n} = V_{max}\frac{C_{n-1}^l}{K^l + C_{n-1}^l}
 \end{cases}
 $.

Като заместим $V_n$ в уравнението за $C_{n+1}$ получаваме едно уравнение за $C$:

$\mathbf{C_{n+1} = C_n - \beta C_n V_{max}\frac{C_{n-1}^l}{K^l + C_{n-1}^l} + m}$

```python
def s(C, K=1, l=0):
    return C**l / (K**l + C**l)
```

```python
C = np.arange(-100, 100)
f, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True)
ax1.plot(C, s(C, K=1, l=0))
ax2.plot(C, s(C, K=1, l=1))
ax3.plot(C, s(C, K=1, l=2))
plt.show()
```

```python
C = np.arange(-50, 100)
f, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True)
ax1.plot(C, s(C, K=82, l=1))
ax2.plot(C, s(C, K=82, l=2))
ax3.plot(C, s(C, K=82, l=3))
plt.show()
```

```python
def model(iterations, K=None, l=None, beta=None, m=None, v_max=None, c0=None):
    c_current = c0
    v_current = v_max * s(c_current, K=K, l=l)
    result = [[c_current, v_current]]
    for _ in range(iterations):
        v_next = v_max * s(c_current, K=K, l=l)
        c_next = c_current - beta * c_current * v_current + m
        result.append([c_next, v_next])
        c_current = c_next
        v_current = v_next
    return result
```

```python
result = model(100, K=1, l=1, beta=1, m=2, v_max=2, c0=2)
plot_model(result)
```
